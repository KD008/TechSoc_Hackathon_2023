```python
import warnings

import pandas as pd

from sdv.evaluation import evaluate
from sdv.metrics.tabular import CSTest, KSTest, LogisticDetection, SVCDetection
from sdv.tabular import CTGAN, TVAE, CopulaGAN, GaussianCopula

warnings.filterwarnings("ignore")
```

```python
import numpy as np
real_data = pd.read_csv("/content/PS1C_train.csv")

print(real_data.shape)
real_data.head()
```

```python
real_data.drop(['Index'], axis=1, inplace=True)
```

```python
model_tvae=TVAE()
```

```python
real_sub_data = pd.read_csv("/content/PS1C_test.csv")
```

```python
real_sub_data.shape[0]
```

```python
data=real_data.copy()
X = data.iloc[:, :-1]
y=data.iloc[:, -1]
```

```python
no_ones=0
no_zeros=0
for i in list(y):
  if i==1:
    no_ones=no_ones+1
  else:
    no_zeros=no_zeros+1
print(no_ones,no_zeros)
```

```python
# print("Before OverSampling, counts of label '1': {}".format(sum(y_train == 1)))
# print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train == 0)))

# import SMOTE module from imblearn library
# pip install imblearn (if you don't have imblearn in your system)
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 2)
Xs, ys = sm.fit_resample(X, y.ravel())

print('After OverSampling, the shape of train_X: {}'.format(Xs.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(ys.shape))

print("After OverSampling, counts of label '1': {}".format(sum(ys == 1)))
print("After OverSampling, counts of label '0': {}".format(sum(ys == 0)))
```

```python
pd.DataFrame(ys)
```

```python
ys=pd.DataFrame(ys)
ys
```

```python
frames = [Xs, ys]
  
result = pd.concat(frames,axis=1, join='inner')
result.rename(columns = {0:'Target'}, inplace = True)
```

```python
pd.pandas.set_option('display.max_columns', None)

result.head(15)
```

```python
model_tvae.fit(result)
```

```python
new_data_tvae = model_tvae.sample(real_sub_data.shape[0])
```

```python
data1=new_data_tvae.copy()
X1 = data1.iloc[:, :-1]
y1=data1.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(X1, y1, test_size = .3, random_state = 10)
```

```python

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 10)
classifier.fit(X1, y1)
```

```python
indices=real_sub_data.loc[:,'Index']
real_sub_data.drop(['Index'], axis=1, inplace=True)
```

```python
y_pred_lr = classifier.predict(real_sub_data)
```

```python
y_pred=pd.DataFrame(y_pred_lr)
```

```python
frames = [indices, y_pred]
  
pred_result = pd.concat(frames,axis=1, join='inner')
```

```python
y_pred.to_csv("1C_subm.csv")
```

```python

from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test_lr, y_pred_lr))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test_lr,y_pred_lr))
```
